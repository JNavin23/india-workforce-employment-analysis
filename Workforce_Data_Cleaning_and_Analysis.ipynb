Data Clean

Rename Columns

```python
work.columns = work.columns.str.strip().str.lower().str.replace(" ","_").str.replace("-","_").str.replace("/","_")
```

```python
work['area_name'] = work['area_name'].str.strip().str.title()
work['religious_community'] = work['religious_community'].str.strip().str.title()
```

Check for Duplicates

```python
work.duplicated().sum()
```

Drop Columns

```python
cols_to_drop = [
'mw_available_total', 'mw_available_males', 'mw_available_females',
'nw_available_total', 'nw_available_males', 'nw_available_females',
'population_males', 'population_females',
'main_workers_male', 'main_workers_female',
'marginal_workers_male', 'marginal_workers_female',
'non_workers_male', 'non_workers_female'
]
work_cleaned = work.drop(columns=cols_to_drop)
```
